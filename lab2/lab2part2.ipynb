```python
import nltk
nltk.download('popular')
nltk.download('all')
```

```python
from nltk.corpus import reuters

text = reuters.raw(categories='cocoa')
print(text)
```

```python
txt = reuters.fileids(categories='cocoa')
document_id = txt[2]

print(reuters.raw(document_id))
```

```python
senteces = reuters.sents(txt[2])
senteces = [ ' '.join(sent) for sent in senteces]
num = len(senteces)
print(senteces[1:num-1])
```

```python
pos_tagged = nltk.pos_tag(nltk.word_tokenize(reuters.raw(document_id)))
verbs = [(word, tag) for word, tag in pos_tagged if any(verb_tag in tag for verb_tag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'])]
verbs_freq = nltk.FreqDist([word for word, tag in verbs])
print(verbs_freq.most_common(8))

import pandas as pd
df = pd.DataFrame(verbs_freq.most_common(8), columns=['word', 'count']).T
df
```
